```python
covid_19['Active_cases']=covid_19['Confirmed']-(covid_19['Deaths']+covid_19['Cured'])
```

```python
covid_19.rename(columns={'State/UnionTerritory':'Place'},inplace=True)
```

```python
covid_19.columns
```

```python
unique_places = covid_19['Place'].unique()
print(unique_places)
```

```python
covid_19['Place'].replace({'Maharashtra***':'Maharashtra','Bihar****':'Bihar','Madhya Pradesh***':'Madhya Pradesh'},inplace=True)
```

```python
covid_19['Place'].unique()
```

```python
statewise_data = pd.pivot_table(covid_19, values=['Confirmed', 'Deaths', 'Cured'], index='Place', aggfunc='max')
```

```python
print(statewise_data)
```

```python
statewise_data['Recovery Rate']=statewise_data['Cured']*100/statewise_data['Confirmed']
statewise_data['Mortality Rate']=statewise_data['Deaths']*100/statewise_data['Confirmed']
```

```python
statewise_data.sort_values(by="Mortality Rate",ascending=False)
```

```python
statewise_data.style.background_gradient(cmap='viridis')
```

```python
top10=covid_19.groupby('Place')
top10
```

```python
top10States=top10.max()[['Active_cases','Cured','Deaths']].sort_values(by=['Active_cases'],ascending=False)
#here the place columns will remain as index if we donot put reset index 
top10States
```

```python
top10States=top10.max()[['Active_cases','Cured','Deaths']].sort_values(by=['Active_cases'],ascending=False).reset_index()
#here the place columns will remain as index if we donot put reset index 
```

```python
top10States
```

```python
top10States.head(10)
```

```python
top10States.columns
```

```python
fig=plt.figure(figsize=(16,10))
plt.title("Top 10 states with most Active Cases  in India",size=25)
sns.barplot(x='Place', y='Active_cases', data=top10States.iloc[:10],palette='plasma')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
top10deaths=top10States.sort_values(by=['Deaths'],ascending=False)
print(top10deaths)
fig=plt.figure(figsize=(12,8))
plt.title("Top 10 states with most Death Cases in India",size=25)
sns.barplot(x='Place', y='Deaths', data=top10deaths.iloc[:10],palette='plasma')
plt.xlabel("States")
plt.ylabel("Total Deaths")
plt.show()
```

```python
top10cured=top10States.sort_values(by=['Cured'],ascending=False)
print(top10cured)
fig=plt.figure(figsize=(13,6))
plt.title("Top 10 states with most Cured Cases in India",size=20)
sns.barplot(x='Place', y='Cured', data=top10cured.iloc[:10],palette='viridis')
plt.xlabel("States")
plt.ylabel("Total Cured")
plt.show()
```

```python
max_vacc = state_vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total Individuals Vaccinated')
max_vacc = max_vacc.sort_values(by='Total Individuals Vaccinated', ascending=False).head(5) 
fig = plt.figure(figsize=(12, 5))
plt.title("Top 5 Vaccinated States in India", size=16)
sns.barplot(data=max_vacc.reset_index(), x='State', y='Total Individuals Vaccinated', linewidth=2, edgecolor='black')  # Use reset_index to include 'State' as a column
plt.xlabel('State')
plt.ylabel('Total Individuals Vaccinated')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
#the .to_frame method converts a Pandas Series into a DataFrame with a specific column name
```

```python
max_vacc = state_vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total Individuals Vaccinated')
max_vacc = max_vacc.sort_values(by='Total Individuals Vaccinated', ascending=False).reset_index()
fig = plt.figure(figsize=(12, 5))
plt.title("Least 5 Vaccinated States in India", size=16)
sns.barplot(data=max_vacc.tail(5), x='State', y='Total Individuals Vaccinated', linewidth=2, edgecolor='black')  # Use reset_index to include 'State' as a column
plt.xlabel('State')
plt.ylabel('Total Individuals Vaccinated') 
plt.show()
```
